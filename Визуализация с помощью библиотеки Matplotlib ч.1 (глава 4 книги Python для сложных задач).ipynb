#### Настройка графика: цвета и стили линий

```python
# в функции plot используем ключевое слово color
plt.plot(x, np.sin(x - 0), color= 'blue') # по названию цвета
plt.plot(x, np.sin(x - 1), color= 'g') # по краткому коду цвета (rgbcmyk)
plt.plot(x, np.sin(x - 2), color= '0.75') # шкала оттенков серого цвета от 0 до 1
plt.plot(x, np.sin(x - 3), color= '#FFDD44') # 16-ричный код цвета (RRGGBB от 00 до FF)
plt.plot(x, np.sin(x - 4), color= (1.0, 0.2, 0.3)) # кортеж RGB, значения от 0 до 1
plt.plot(x, np.sin(x - 5), color= 'chartreuse') # по названию цвета HTML
```

Что такое RGB?

Аббревиатура RGB означает названия трех цветов, использующихся для вывода на экран цветного изображения: Red (красный), Green (зеленый), Blue (синий).

Как формируется цвет RGB?

Цвет на экране монитора формируется при объединении лучей трех основных цветов - красного, зеленого и синего. Если интенсивность каждого из них достигает 100%, то получается белый цвет. Отсутствие всех трех цветов дает черный цвет.

Таким образом, любой цвет, который мы видим на экране, можно описать тремя числами, обозначающими яркость красной, зеленой и синей цветовых составляющих в цифровом диапазоне от 0 до 255. Графические программы позволяют комбинировать требуемый RGB-цвет из 256 оттенков красного, 256 оттенков зеленого и 256 оттенков синего. Итого получается 256 х 256 х 256 = 16,7 миллионов цветов.

###### Непрерывные погрешности


```python
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
model = lambda x: x*np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)
```

```python
gp = GaussianProcessRegressor(random_state=100)
```

```python
gp.fit(xdata[:, np.newaxis], ydata)
```

```python
xfit = np.linspace(0,10,1000)
yfit, STDV = gp.predict(xfit[:, np.newaxis], return_std=True)
```

```python
dyfit = 2 * np.sqrt(STDV)
```

```python
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                color='gray', alpha = 0.2)
```

#### Графики плотности и контурные графики

plt.contourf - для контурных графиков с заполнением

plt.imshow - для отображения картинок

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

Визуализация трехмерной функции z = f (x, y)

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y*x) * np.cos(x)
```

np.meshgrid - формирует двумерные координатные сетки из одномерных массивов

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
plt.contour(X,Y,Z, colors='black') # пунктиром отмечены отрицательные значения
```

```python
# используем сплошной контур
plt.contourf(X,Y,Z, 50, cmap='RdGy')
plt.colorbar();
```

```python
# если нужно избавиться от "пятнистости"
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image');
```
